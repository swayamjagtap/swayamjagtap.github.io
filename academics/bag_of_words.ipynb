#### imports

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

#### Step 1 : Define a sample corpus of text documents

```python
corpus = [
    "I love coding in python",
    "Python is great for coding",
    "I love building NLP models"
]
```

#### Step 2 : Initialize the CounterVectorizer

```python
vectorizer = CountVectorizer()
```

#### Step 3 : `Fit` the model & `Transform` the text into sparse matrix

```python
bow_matrix = vectorizer.fit_transform(corpus)
bow_matrix
```

#### Step 4 : Extract vocabulary names(unique words)

```python
vocabulary = vectorizer.get_feature_names_out()
vocabulary
```

#### Step 5 : Convert the matrix to an array and display it as a readable DataFrame

```python
bow_df = pd.DataFrame(bow_matrix.toarray(), columns = vocabulary)

print("Vocabulary List")
print(vocabulary)
print("\nBag of Words matrix:")
print(bow_df)
```
