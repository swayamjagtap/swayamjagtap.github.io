# Step 1: load all `.txt` files from a folder

```python
folder_path = r"C:\Users\IT2\Downloads\swayam"
files = glob.glob(folder_path + "/*.txt")

corpus = [open(f, encoding = "utf-8").read() for f in files]
print(f"Loaded {len(corpus)} documents.")
```

# Step 2: Tokenization and Normalization

```python
tokens = []
for doc in corpus:
    words = nltk.word_tokenize(doc.lower())
    tokens.extend([w for w in words if w.isalpha()])
```

# Step 3: Count word frequencies

```python
freq = Counter(tokens)
print("Top 20 words in corpus: ")
for word, count in freq.most_common(20):
    print(word, ":", count)
```
