## AIM: To implement stemming and lemmatization using NLP libraries

### 1. using NLTK library

#### Step 1: Import libraries

```python
import nltk
# nltk.download('punkt')
# nltk.download('wordnet')
# nltk.download('omw-1.4')
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

#### Step 2: Initialize `Stemmer` and `Lemmatizer`

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

#### Step 3: Create a list of words and perform `stemming` and `lemmatization`

```python
words = ['running','flies','studying','better','leaves']

print("Swayam Jagtap - 16")
print(f"{'Original word':<15} | {'Stemmed(Porter)':<18} | {'Lemmatized(WordNet)':<20}")
print('-'*60)

for w in words:
    stemmed = stemmer.stem(w)
    lemmatized = lemmatizer.lemmatize(w)

    print(f"{w:<15} | {stemmed:<18} | {lemmatized:<20}")
```

### 2. using spaCy library

```python
import spacy
nlp = spacy.load('en_core_web_sm')
text = 'The children are dancing and laughing in enjoyment'
doc = nlp(text)
print('Swayam Jagtap - 16')
print('Word\t\tlemma')
print('-'*25)

for token in doc:
    print(f"{token.text:12} | {token.lemma_}")          #token.lemma_  underscore is important
```
